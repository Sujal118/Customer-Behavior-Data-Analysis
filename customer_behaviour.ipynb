```python
import pandas as pd
df=pd.read_csv('customer_shopping_behavior.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))

#we are filling the null places of rating by grouping them category wise and median. category bcoz we r avoiding rating of clothes to shoe are anything else and meadian bcoz it is safer not that sensetive as compare to mean
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
 # createing a column age_grp
labels = ['young adult','adult','middle aged','senior']
df['age_grp']=pd.qcut(df['age'], q=4, labels=labels)
df.head(200)
```

```python
#create a purchase frequency days

frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
df.head(20)
```

```python
(df['discount_applied']==df['promo_code_used']).all()
df=df.drop('promo_code_used',axis=1)
```
